## 针对原始切分文件，统计不同词条所属标签的数量分布情况

- 目标：直观展现不同标签的数量差异情况


```python
# 将所有出现过的标签进行统计
import csv
import pandas as pd

label_list = []

df = pd.read_csv('./drive/MyDrive/Database/split/split-ownthink-2.csv', usecols=['实体','属性', '值'], index_col=False)
df.reset_index(drop=True, inplace=True)

for i in range(len(df)):
    if df.loc[i, '属性']=='标签':
        if df.loc[i, '值'] not in label_list:
            label_list.append(df.loc[i, '值'])

print('label_list: ', label_list)
```

```python
dict_label = {}

for i in label_list:
    dict_label[i] = 0

for j in range(len(df)):
    if df.loc[j, '属性']=='标签':
        label = df.loc[j, '值']
        dict_label[label] = dict_label[label] + 1

print(dict_label)
```

```python
sorted_tuples = sorted(dict_label.items(), key=lambda item: item[1])
sorted_dict = {k: v for k, v in sorted_tuples[::-1]}
print(sorted_dict)
```

```python
x = []
y = []
for i in sorted_dict:
    x.append(i)
    y.append(sorted_dict[i])

print(y)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

zhfont = mpl.font_manager.FontProperties(fname='/usr/share/fonts/truetype/liberation/simhei.ttf')
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
plt.rcParams['axes.unicode_minus'] =False
fig = plt.figure(figsize=(15, 10))
plt.bar(x[:20] , y[:20], align='center') # A bar chart
plt.title(u'知识图谱标签分布情况统计', color='r', fontproperties=zhfont)
plt.xlabel(u'标签名--降序', fontproperties=zhfont)
plt.ylabel(u'数量', fontproperties=zhfont)
plt.xticks(rotation=45, fontproperties=zhfont)
fig.savefig('figure-4.png', dpi=100)
plt.show()
```
