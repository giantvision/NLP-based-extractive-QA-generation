```python
from bs4 import BeautifulSoup
import urllib.request as urllib2
from lxml import etree, html
import requests
from pyhanlp import *

def convertToSimplifiedChinese(text):
    return HanLP.convertToSimplifiedChinese(text)
```

```python
# Wiki 进行 query_word 的基本语义爬取
page = requests.get('https://zh.wikipedia.org/wiki/惊蛰')
tree = html.fromstring(page.content)

text = tree.xpath('//*[@id="mw-content-text"]/div[1]/p[1]//text()')
# print('text: ', text)
string = ''
for i in text:
    string = string + i
print("Content: ", string)
```

```python
def SpiderWiki(query_list, name_list):
    for k in name_list:
        for i in query_list:
            print(i)
            # Wiki 进行 query_word 的基本语义爬取
            page = requests.get(f'https://zh.wikipedia.org/wiki/{i}'.format(i))
            tree = html.fromstring(page.content)
            text = ''
            for j in range(3):
                k = j + 1
                text = tree.xpath(f'//*[@id="mw-content-text"]/div[1]/p[{k}]//text()'.format(k))
                string = ''
                for m in text:
                    string = string + m
                if string == '':
                    pass
                else:
                    Content_switch = convertToSimplifiedChinese(string)
                    with open('./Festivals-wiki.txt', 'a') as f:
                        f.write(Content_switch)
                    print(Content_switch)
    info_message = 'The spider is finished!'
    return info_message
```

```python
# 传统文化相关
# 9、传统节日
TraditionalFestivals_list = ['小年', '孟兰节', '立春', '惊蛰']

SpiderWiki(TraditionalFestivals_list)
```

```python
import re
def SpiderKmcha(query_dict):
    for cur_key in query_dict:
        print('Current Query Dict: ', cur_key)
        string_list = []
        for i in query_dict[cur_key]:
            print('Query Word: ', i)
            # Wiki 进行 query_word 的基本语义爬取
            page = requests.get(f'https://kmcha.com/similar/{i}'.format(i))
            tree = html.fromstring(page.content)
            text = tree.xpath('/html/body/div[1]/div[2]/div[4]/p[3]//text()')
            text_2 = tree.xpath('/html/body/div[1]/div[2]/div[4]/div[1]/p//text()')
            string_text = ''
            for j in text:
                string_text = string_text + j
            string_text = string_text.strip()
            string_text = re.sub(r'([0-9]+\.)', r'', string_text)
            string_text = re.sub(r'(\[.*?])', r'', string_text)
            # print('string_text: ', string_text)
            string_text_2 = ''
            for k in text_2:
                string_text_2 = string_text_2 + k

            string_text_2 = string_text_2.strip()
            string_text_2 = string_text_2.replace('\n',"")
            string_text_2 = re.sub(r'(\[.*?])', r'', string_text_2)
            
            # print('string_text_2: ', string_text_2)
            string_concat = i + '，' + string_text + '。' + string_text_2
            string_list.append(string_concat)
            print('string_concat: ', string_concat)
        with open(f'./drive/MyDrive/Database/extract_results/20220622/kmcha/{cur_key}-kmcha.txt'.format(cur_key), 'a', encoding='utf-8-sig') as f:
            for i in string_list:
                f.write(i)
                f.write('\n')
```
