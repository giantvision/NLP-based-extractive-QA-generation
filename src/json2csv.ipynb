```python
import pandas as pd
import json

filename = './drive/MyDrive/Database/results/20220418/Chinese_culture.json'


with open(filename, encoding='utf-8-sig') as json_file:
    data = json.load(json_file)
    print(data)
```

```python
df = pd.DataFrame(data)
```

```python
list_data = []

for i in data:
    for qa in i['qa_pair']:
        list_data.append([i['_id'], i['text'], qa['q'], qa['a'], qa['label']])

new_df = pd.DataFrame(list_data, columns=['_id', 'text', 'q', 'a', 'label'])
print('new_df: ', new_df.head())

```

```python
csv_file = new_df.to_csv('file_csv.csv', sep='\t', index=False)
```
