```python
import os

def remove_empty_lines(filename):

    if not os.path.isfile(filename):
        print("{} does not exist ".format(filename))
        return

    with open(filename) as filehandle:
        lines = filehandle.readlines()
        # print('lines: ', lines)
        with open('drive/MyDrive/Database/test_data/CollectedInfo/NewPhy.txt', 'w') as filehandle:
            lines = filter(lambda x: x.strip(), lines)
            print(lines)
            filehandle.writelines(lines)

remove_empty_lines('drive/MyDrive/Database/test_data/CollectedInfo/Phy.txt')
```

```python
import pandas as pd
import glob
import os

path = r'drive/MyDrive/Database/test_data/CollectedInfo/CSV' # use your path

all_files = glob.glob(path + "/*.csv")

for filename in all_files:
    df = pd.read_csv(filename, on_bad_lines='skip')
    df.columns = ['line']
    fname = filename.split('/')[-1:][0]
    fname = os.path.splitext(fname)[0]
    df.to_csv('./drive/MyDrive/Database/test_data/CollectedInfo/add_cols/'+fname+ '.csv', encoding='utf-8-sig')
```
