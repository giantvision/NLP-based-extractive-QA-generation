## 直接从网页链接进行信息提取

```python
from bs4 import BeautifulSoup
import urllib.request as urllib2
from lxml import etree, html
import requests

page = requests.get('https://en.wikipedia.org/wiki/Glossary_of_biology')
tree = html.fromstring(page.content)
#This will create a list of buyers:
query_word = tree.xpath('//*[@id!=""]/dfn/a/text()')
print('Title: ', query_word)
```
