```python
import numpy as np


chosenSize = 100
totalSize = 0
count_num = 0

for j in data_list:
    totalSize += j
print('totalSize: ', totalSize)

for i in data_list[:-1]:
    chosen_num = np.rint(i*chosenSize/totalSize)
    if count_num < 100 and chosen_num+count_num >100:
        chosen_num = 100 - count_num
        print('******')
        break
    count_num += chosen_num 
    print('chosen_num: ', int(chosen_num))

print('count_num: ', int(count_num))

last_chosen = chosenSize - count_num
print('last_chosen: ', int(last_chosen))
```
