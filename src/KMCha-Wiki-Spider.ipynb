# 对KMCha数据进行定向爬取工作

```python
# 传统文化相关
# 9、传统节日
TraditionalFestivals_list = ['小年', '孟兰节', '立春', '惊蛰', '立夏', '火把节', '腊八节', '大雪', '七夕', '霜降', '大暑', 
                        '小暑', '中元节', '春节', '地藏节', '立秋', '小雪', '鬼节', '立冬', '白露', '秋分', '芒种', 
                        '龙抬头', '大寒', '小寒', '处暑', '夏至', '祭祖节', '小满', '雨水', '春分', '除夕', '寒露', 
                        '中秋节', '谷雨', '元宵节', '阔时节', '冬至', '清明节', '寒食节', '重阳节', '翻经节', '清明', '姑姑节']
```

```python
import re
```

```python
string_list = []
for i in TraditionalFestivals_list:
    print(i)
    # Wiki 进行 query_word 的基本语义爬取
    page = requests.get(f'https://kmcha.com/similar/{i}'.format(i))
    tree = html.fromstring(page.content)
    text = tree.xpath('/html/body/div[1]/div[2]/div[4]/p[3]//text()')
    text_2 = tree.xpath('/html/body/div[1]/div[2]/div[4]/div[1]/p//text()')
    string_text = ''
    for j in text:
        string_text = string_text + j
    string_text = string_text.strip()
    string_text = re.sub(r'([0-9]+\.)', r'', string_text)
    string_text = re.sub(r'(\[.*?])', r'', string_text)
    # print('string_text: ', string_text)
    string_text_2 = ''
    for k in text_2:
        string_text_2 = string_text_2 + k

    string_text_2 = string_text_2.strip()
    string_text_2 = string_text_2.replace('\n',"")
    string_text_2 = re.sub(r'(\[.*?])', r'', string_text_2)
    
    # print('string_text_2: ', string_text_2)
    string_concat = i + '，' + string_text + '。' + string_text_2
    string_list.append(string_concat)
    print('string_concat: ', string_concat)

with open('./drive/MyDrive/Database/extract_results/20220615/kmcha/Festivals-kmcha.txt', 'a', encoding='utf-8-sig') as f:
    for i in string_list:
        f.write(i)
        f.write('\n')
```

## 测试爬取过程中的异常问题

- 原因总结：

   由于爬取的原文本中存在 '\n' 的换行符，从而导致写入过程出现异常，此时应该注意 string.strip() 只能处理 string 的开头和结尾的 '\n' 或者 '\t' 的异常，对于此处的处于 string 字符串中间的情况，则应该使用 string.replace() 进行操作，可解决该问题。

```python
page = requests.get('https://kmcha.com/similar/腊八节')
tree = html.fromstring(page.content)

text = tree.xpath('/html/body/div[1]/div[2]/div[4]/p[3]//text()')
text_2 = tree.xpath('/html/body/div[1]/div[2]/div[4]/div[1]/p//text()')
print('text: ', text)
print('text_2: ', text_2)

string_text = ''
for j in text:
    string_text = string_text + j
string_text = string_text.strip()
string_text = re.sub(r'([0-9]+\.)', r'', string_text)
# string_text = re.sub(r'(\[.*?])', r'', string_text)
print('string_text: ', string_text)

string_text_2 = ''
for k in text_2:
    string_text_2 = string_text_2 + k
string_text_2 = string_text_2.strip('\n')
string_text_2 = string_text_2.replace('\n',"")
string_text_2 = re.sub(r'(\[.*?])', r'', string_text_2)

print('string_text_2: ', string_text_2)


# string_concat = i + '，' + string_text + '。' + string_text_2
# string_list.append(string_concat)
# print('string_concat: ', string_concat)
```
